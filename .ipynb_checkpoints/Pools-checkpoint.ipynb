```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
poolData = pd.read_csv('https://data.wprdc.org/datastore/dump/5cc254fe-2cbd-4912-9f44-2f95f0beea9apoolData')

poolDict = dict()
```

```python
for i in range(len(poolData)):
    item = poolData.loc[i, 'neighborhood']
    
    if item in poolDict:
        poolDict[item] += 1
    else:
        poolDict[item] = 1
        
print(poolDict)
```
